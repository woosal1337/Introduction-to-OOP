### Decorators

````python
# @property 
# @staticmethods
# @classmethods

import random

CUSTOM_LIST = [
    "Fuat Beser",
    "Omer Cengiz",
    "Mert Cobanoglu",
    "Vusal Ismayilov",
    "Berkay Ermis"
]

class DecoratorClass:
    
    # Custom class variable
    _CLASS_VAR = [
        "CLASS",
        "VARIABLE"
    ]
    
    _CLASS_VAR_2 = [
        "ANOTHER",
        "CLASS",
        "VAR"
    ]
    
    # Functions
    def __init__(self, name):
        # Class specific declaration
        self._name = name
        
    def get_name(self):
        
        return self._name
    
    def set_name(self, name):
        
        self._name = name
        
    @property
    def get_name_decorated(self):
        
        return self._name
    
    @get_name_decorated.setter
    def set_name_decorated(self, name):
        
        self._name = name
        
    @staticmethod
    def random_user():
        
        return random.choice(CUSTOM_LIST)
    
    @classmethod
    def class_met(cls):
        
        return cls._CLASS_VAR
    
    @classmethod
    def class_met_2(cls):
        
        return cls._CLASS_VAR_2
````

# Card Class
### https://en.wikipedia.org/wiki/Standard_52-card_deck

````python
class Cards:
    def __init__(self):
        self._cardList = []
    
    def addCard(self, cardName, cardNumber = -1):
        newCard = [cardName, cardNumber]
        
        if newCard not in self._cardList:
            self._cardList.append(newCard)
            return newCard
        else:
            return "Card is already in the list!"
    
    def removeCard(self, cardName, cardNumber = -1):
        manipulatedInput = [cardName, cardNumber]
        
        if manipulatedInput in self._cardList:
            self._cardList.pop(self._cardList.index(manipulatedInput))
            return manipulatedInput
        else:
            return "Card does not exist!"
    
    def showCardList(self):
        return self._cardList
````

````python
# You don't have to invoke it directly

class Length:

    __metric = {"mm" : 0.001, "cm" : 0.01, "m" : 1, "km" : 1000,
                "in" : 0.0254, "ft" : 0.3048, "yd" : 0.9144,
                "mi" : 1609.344 }
    
    def __init__(self, value, unit = "m" ):
        self.value = value
        self.unit = unit
    
    def Converse2Metres(self):
        return self.value * Length.__metric[self.unit]
    
    def __add__(self, other):
        l = self.Converse2Metres() + other.Converse2Metres()
        return Length(l / Length.__metric[self.unit], self.unit )
    
    def __str__(self):
        return str(self.Converse2Metres())
    
    # https://www.pythontutorial.net/python-oop/python-__repr__/
    def __repr__(self):
        return "Length(" + str(self.value) + ", '" + self.unit + "')"
        
    def __len__(self):
        print(f"The metric dictionary length is {len(self.__metric)}.")
        return len(self.__metric)
        
if __name__ == "__main__":
    x = Length(4)
    print(len(x))
    
    print(x)
    y = eval(repr(x))

    z = Length(4.5, "yd") + Length(1)
    print(repr(z))
    print(z)
````

````python
# __str__ vs __repr__

class Person:
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

    def __repr__(self):
        return f'Person("{self.first_name}","{self.last_name}",{self.age})'

    def __str__(self):
        return f'({self.first_name},{self.last_name},{self.age})'


person = Person('John', 'Doe', 25)
# use str()
print(person)

# use repr()
print(repr(person))
````

````python
class TriangleArea:
    
    def __call__(self, a, b, c):
        p = (a + b + c) / 2
        result = (p * (p - a) * (p - b) * (p - c)) ** 0.5
        return result
    
    
area = TriangleArea()

print(area(3, 4, 5))
````

````python
# __call__ takikng arguments
class StraightLines():
    
    def __init__(self, m, c):
        self.slope = m
        self.y_intercept = c
        
    def __call__(self, x):
        return self.slope * x + self.y_intercept
    
line = StraightLines(0.4, 3)

for x in range(-5, 6):
    print(x, line(x))
````

# Overloading Operators

````python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0},{1})".format(self.x, self.y)

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)


p1 = Point(1, 2)
p2 = Point(2, 3)

print(p1+p2)
````

````python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0},{1})".format(self.x, self.y)

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)
    
    def __sub__(self, other):
        x = self.x - other.x
        y = self.y - other.y
        return Point(x,y)
    
    def __mul__(self, other):
        x = self.x * other.x
        y = self.y * other.y
        return Point(x,y)
    
    def __pow__(self, other):
        x = self.x ** other.x
        y = self.y ** other.y
        return Point(x,y)
    
    def __truediv__(self, other):
        x = self.x / other.x
        y = self.y / other.y
        return Point(x,y)
    
    def __floordiv__(self, other):
        x = self.x // other.x
        y = self.y // other.y
        return Point(x,y)
    
    def __mod__(self, other):
        x = self.x % other.x
        y = self.y % other.y
        return Point(x,y)
    
p1 = Point(1, 2)
p2 = Point(2, 3)

print(p1-p2)
print(p1*p2)
print(p1 ** p2)
print(p1/p2)
print(p1//p2)
print(p1%p2)
````

# Overloading Comparison Operators

````python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0},{1})".format(self.x, self.y)

    def __lt__(self, other):
        self_mag = (self.x ** 2) + (self.y ** 2)
        other_mag = (other.x ** 2) + (other.y ** 2)
        return self_mag < other_mag


    def __le__(self, other):
        self_mag = (self.x ** 2) + (self.y ** 2)
        other_mag = (other.x ** 2) + (other.y ** 2)
        return self_mag < other_mag
    
p1 = Point(1,1)
p2 = Point(-2,-3)
p3 = Point(1,-1)


print(p1<p2)
````

# Abstract Classes and Methods

````python
from abc import ABC, abstractmethod
 
class Polygon(ABC):
 
    @abstractmethod
    def noofsides(self):
        pass
 
class Triangle(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 3 sides")
 
class Pentagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 5 sides")
 
class Hexagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 6 sides")
 
class Quadrilateral(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 4 sides")
 
# Driver code
R = Triangle()
R.noofsides()
 
K = Quadrilateral()
K.noofsides()
 
R = Pentagon()
R.noofsides()
 
K = Hexagon()
K.noofsides()
````

# Polymorphism

````python
class Document:
    def __init__(self, name):
        self.name = name
    
    def show(self):
        raise NotImplementedError("Subclass must implement abstract method")
    
class Pdf(Document):
    def show(self):
        return 'Show pdf contents!'

class Word(Document):
    def show(self):
        return 'Show word contents!'
    
documents = [Pdf('Document1'),
Pdf('Document2'),
Word('Document3')]

for document in documents:
    print(document.name + ': ' + document.show())
````

````python
class Document:
    def __init__(self, name):
        self.name = name
    
    def show(self):
        raise NotImplementedError("Subclass must implement abstract method")
    
class Pdf(Document):
    def show(self):
        return 'Show pdf contents!'

class Word(Document):
    def show(self):
        return 'Show word contents!'
    
class Excel(Document):
    def show(self):
        pass
    
documents = Excel("Untitled.xlsx")

documents.show()
````

````python
class Car:
    def __init__(self, name):
        self.name = name
    
    def drive(self):
        raise NotImplementedError("Subclass must implement abstract method")

    def stop(self):
        raise NotImplementedError("Subclass must implement abstract method")
    
class Sportscar(Car):
    def drive(self):
        return 'Sportscar driving!'

    def stop(self):
        return 'Sportscar braking!'
    
class Truck(Car):
    def drive(self):
        return 'Truck driving slowly because heavily loaded.'

    def stop(self):
        return 'Truck braking!'

cars = [Truck('Bananatruck'),
Truck('Orangetruck'),
Sportscar('Z3')]

for car in cars:
    print(car.name + ': ' + car.drive())
````
