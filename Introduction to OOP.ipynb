```python
# Classic and most known class examples:

# Implementation of class goes here:
class Person:
    def __init__(self, firstName, lastName, gender):
        self.firstName = firstName
        self.lastName = lastName
        self.gender = gender
    
    def get_name(self):
        return self.firstName
    
# Use of the implemented classes go here:
John = Person("John", "Doe", "male")
John.get_name()
```

```python
# Modifying the previous class:
class Person:
    def __init__(self, firstName, lastName, gender):
        self.firstName = firstName
        self.lastName = lastName
        self.gender = gender
    
    def getName(self):
        return self.firstName
    
    def getLastName(self):
        return self.lastName
    
    def getGender(self):
        return self.gender
    
Vusal = Person("Vusal", "Ismayilov", "male")
Vusal.getName()
```
