### Inheritance
Every class in Python is derived from the object class.

```python
# Inheritance:
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        FirstClass.__init__(self)
        
second_class = SecondClass()
```

### Multiple inheritance

```python
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        FirstClass.__init__(self)
        
class ThirdClass(FirstClass):
    def __init__(self):
        print("Third Class")
        FirstClass.__init__(self)
        
class FourthClass(SecondClass, ThirdClass):
    def __init__(self):
        print("Fourth Class")
        SecondClass.__init__(self)
        ThirdClass.__init__(self)
        
fourth_class = FourthClass()
```

### Super


```python
# super()
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        super().__init__() #super(SecondClass, self).__init__()
        
second_class = SecondClass()
```

```python
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        super().__init__()
        
class ThirdClass(FirstClass):
    def __init__(self):
        print("Third Class")
        super().__init__()
        
class FourthClass(SecondClass, ThirdClass):
    def __init__(self):
        print("Fourth Class")
        super().__init__()
        
fourth_class = FourthClass()
```

### Decorators

```python
# @property 
# @staticmethods
# @classmethods

import random

CUSTOM_LIST = [
    "Fuat Beser",
    "Omer Cengiz",
    "Mert Cobanoglu",
    "Vusal Ismayilov",
    "Berkay Ermis"
]

class DecoratorClass:
    
    # Custom class variable
    _CLASS_VAR = [
        "CLASS",
        "VARIABLE"
    ]
    
    _CLASS_VAR_2 = [
        "ANOTHER",
        "CLASS",
        "VAR"
    ]
    
    # Functions
    def __init__(self, name):
        # Class specific declaration
        self._name = name
        
    def get_name(self):
        
        return self._name
    
    def set_name(self, name):
        
        self._name = name
        
    @property
    def get_name_decorated(self):
        
        return self._name
    
    @get_name_decorated.setter
    def set_name_decorated(self, name):
        
        self._name = name
        
    @staticmethod
    def random_user():
        
        return random.choice(CUSTOM_LIST)
    
    @classmethod
    def class_met(cls):
        
        return cls._CLASS_VAR
    
    @classmethod
    def class_met_2(cls):
        
        return cls._CLASS_VAR_2
```

```python
myDecClass = DecoratorClass("Umut")
```

```python
# Calling the usual get_na
print(myDecClass.get_name())
```

```python
# Calling @operator function
print(myDecClass.get_name_decorated)
```

```python
# Calling the random function
for i in range(10):
    print(myDecClass.random_user())
```

```python
# Calling @classmethod
print(myDecClass.class_met())
```

```python
print(myDecClass.class_met_2())
```

```python
# Direct access to the variables
print(myDecClass._name)
```

```python
myDecClass.set_name("Vusal")

print(myDecClass.get_name_decorated)
```

```python
# __len__

class DunderClass:
    
    def __init__(self):
        
        self._users = [
            "Vusal",
            "Mert",
            "Fuat",
            "Umut",
            "Gurkan",
            "Ezgi",
            "Kutay",
            "Omer"
        ]
    
    def __len__(self):
        
        return len(self._users)
    
    
dunder_class = DunderClass()
print(len(dunder_class))
```

```python
# __getitem__

class DunderClass:
    
    def __init__(self):
        
        self._users = [
            "Vusal",
            "Mert",
            "Fuat",
            "Umut",
            "Gurkan",
            "Ezgi",
            "Kutay",
            "Omer"
        ]
    
    def __len__(self):
        
        return len(self._users)
    
    def __getitem__(self, index):
        
        if isinstance(index, int):
            
            return self._users.pop(index)
            
        raise TypeError("Asked index value is not an integer!")
            
    
dunder_class = DunderClass()
print("Total users: %d" % len(dunder_class))
print("Excluded member: %s" % dunder_class[0])
print("Total users: %d" % len(dunder_class))
```

# Card Class
### https://en.wikipedia.org/wiki/Standard_52-card_deck

```python
class Cards:
    def __init__(self):
        self._cardList = []
    
    def addCard(self, cardName, cardNumber = -1):
        newCard = [cardName, cardNumber]
        
        if newCard not in self._cardList:
            self._cardList.append(newCard)
            return newCard
        else:
            return "Card is already in the list!"
    
    def removeCard(self, cardName, cardNumber = -1):
        manipulatedInput = [cardName, cardNumber]
        
        if manipulatedInput in self._cardList:
            self._cardList.pop(self._cardList.index(manipulatedInput))
            return manipulatedInput
        else:
            return "Card does not exist!"
    
    def showCardList(self):
        return self._cardList
```

```python
my_cards = Cards()
```

```python
my_cards.addCard("Joker")
```

```python
my_cards.removeCard("Joker")
```

```python
my_cards.showCardList()
```

# `*args **kwargs`

```python
# Summing infinite amount of input arguments / *args 
def addAll(x):
    return sum(x)

addAll(5,5,6,7)
```

```python
def addAll(*x):
    return sum(x)

addAll(2,5,5,5,6,6)
```

```python
addAll(5,6,7,5,75,67,4,32,52,5,352,352,35,235,23)
```

```python
# Tuple
def addAll(*x):
    return x

addAll(2,5,5,5,6,6)
```
