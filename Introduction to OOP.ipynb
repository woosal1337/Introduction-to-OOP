### Inheritance
Every class in Python is derived from the object class.

```python
# Inheritance:
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        FirstClass.__init__(self)
        
second_class = SecondClass()
```

### Multiple inheritance

```python
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        FirstClass.__init__(self)
        
class ThirdClass(FirstClass):
    def __init__(self):
        print("Third Class")
        FirstClass.__init__(self)
        
class FourthClass(SecondClass, ThirdClass):
    def __init__(self):
        print("Fourth Class")
        SecondClass.__init__(self)
        ThirdClass.__init__(self)
        
fourth_class = FourthClass()
```

### Decorators

```python
# @property 
# @staticmethods
# @classmethods

import random

CUSTOM_LIST = [
    "Fuat Beser",
    "Omer Cengiz",
    "Mert Cobanoglu",
    "Vusal Ismayilov"
]

class DecoratorClass:
    
    # Custom class variable
    _CLASS_VAR = [
        "CLASS",
        "VARIABLE"
    ]
    
    _CLASS_VAR_2 = [
        "ANOTHER",
        "CLASS",
        "VAR"
    ]
    
    # Functions
    def __init__(self, name):
        # Class specific declaration
        self._name = name
        
    def get_name(self):
        
        return self._name
    
    def set_name(self, name):
        
        self._name = name
        
    @property
    def get_name_decorated(self):
        
        return self._name
    
    @get_name_decorated.setter
    def set_name_decorated(self, name):
        
        self._name = name
        
    @setter
    def name_setter(self, name):
        
        self._name = name
        
    @staticmethod
    def random_user():
        
        return random.choice(CUSTOM_LIST)
    
    @classmethod
    def class_met(cls):
        
        return cls._CLASS_VAR
    
    @classmethod
    def class_met_2(cls):
        
        return cls._CLASS_VAR_2
```

```python
myDecClass = DecoratorClass("Umut")
```

```python
# Calling the usual get_na
print(myDecClass.get_name())
```

```python
# Calling @operator function
print(myDecClass.get_name_decorated)
```

```python
# Calling the random function
for i in range(10):
    print(myDecClass.random_user())
```

```python
# Calling @classmethod
print(myDecClass.class_met())
```

```python
print(myDecClass.class_met_2())
```
