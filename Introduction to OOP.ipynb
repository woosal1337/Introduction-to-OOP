### Inheritance
Every class in Python is derived from the object class.

```python
# Inheritance:
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        FirstClass.__init__(self)
        
second_class = SecondClass()
```

```python
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        FirstClass.__init__(self)
        
class ThirdClass(FirstClass):
    def __init__(self):
        print("Third Class")
        FirstClass.__init__(self)
        
third_class = ThirdClass()
```

### Multiple inheritance

```python
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        FirstClass.__init__(self)
        
class ThirdClass(FirstClass):
    def __init__(self):
        print("Third Class")
        FirstClass.__init__(self)
        
class FourthClass(SecondClass, ThirdClass):
    def __init__(self):
        print("Fourth Class")
        SecondClass.__init__(self)
        ThirdClass.__init__(self)
        
fourth_class = FourthClass()
```

### Super


```python
# super()
class FirstClass:
    def __init__(self):
        print("First Class")

class SecondClass(FirstClass):
    def __init__(self):
        print("Second Class")
        super().__init__() #super(SecondClass, self).__init__()
        
second_class = SecondClass()
```
